```python
#To read from file
def unsual_option_actv():
    
    today_date = dt.datetime.today().strftime('%m-%d-%Y') # This can be used to read the file
    
    unusual_stock_options_activity = pd.read_csv(f'unusual-stock-options-activity-{today_date}.csv')
    unusual_stock_options_activity = unusual_stock_options_activity.iloc[:-1,:]
    unusual_stock_options_activity.rename(columns={'Exp Date': 'Exp_Date', 'Open Int': 'Open_Int'}, inplace=True)
    unusual_stock_options_activity.IV = unusual_stock_options_activity.IV.str.replace(r',', "")
    unusual_stock_options_activity.IV = unusual_stock_options_activity.IV.str.replace(r'%', "").astype('float')
    
#     publish_one(unusual_stock_options_activity, 'unusual_stock_options_activity')

    unusual_stock_options_activity.Exp_Date = pd.to_datetime(unusual_stock_options_activity.Exp_Date.values).strftime('%Y-%m-%d')
    unusual_stock_options_activity.Time = pd.to_datetime(unusual_stock_options_activity.Time.values)
    
    unusual_stock_options_activity = unusual_stock_options_activity[['Symbol', 'Price', 'Type', 'Strike', 'Exp_Date', 'DTE', 'Volume', 'Open_Int', 'Vol/OI', 'IV','Delta', 'Time']]
    unusual_stock_options_activity = unusual_stock_options_activity[unusual_stock_options_activity.Exp_Date <= exp_date]
    
    print('unusual_stock_options_activity read successfully')
    print('x'*100)
    
    return unusual_stock_options_activity

# df_all =  unsual_option_actv()
# df_all.head(3)
```

```python
# To read from SQL Table
def make_con(cal_date, exp_date):

    conn=pymysql.connect(host='********',
                         port=int(*****),
                         user='*******',
                         passwd='*********',
                         db='Options')
    
    current_date = pd.datetime.today().strftime('%Y-%m-%d') #this can be used instead of attributes above
 

    df_all =pd.read_sql_query(f'SELECT * FROM options.unusual_stock_options_activity WHERE Time = "{cal_date}";',conn)     
    
    df_all= df_all[(df_all.Last >= df_all.Ask)]

    df_all = df_all[['Symbol', 'Price', 'Type', 'Strike', 'Exp_Date', 'DTE', 'Volume', 'Open_Int', 'Vol/OI', 'IV','Delta', 'Time']]
    df_all = df_all[df_all.Exp_Date <= pd.to_datetime(exp_date)]
    
    return df_all

df_all = make_con(current_date, exp_date)
df_all.head(3)
```

```python
def get_yahoo_data(df_all, start_day=1, end_day=0):
    
    

    stocks = df_all.Symbol.unique().tolist()
    
    start = (pd.datetime.now() - BDay(start_day)).strftime('%Y-%m-%d')
    end = (dt.datetime.today() + relativedelta(days=end_day)).strftime('%Y-%m-%d')
    
    new = yf.download(stocks, start, end, auto_adjust=True, progress=True)['Close']
    new = pd.DataFrame(new.T)
    
    new.rename(columns={new.columns[0]: 'Yesterday'}, inplace = True)
    new.reset_index(inplace=True)
    new.rename(columns={'index':'Symbol'}, inplace=True)
#     new = new.iloc[:,:2]
    
    return new
```

```python
new = get_yahoo_data(df_all, 1, 0)
```

```python
def concat_df(df_all, df1):
    
    df2 = df_all[['Symbol', 'Price']]
    df2 = df2.drop_duplicates()
    df2 = df2.sort_values(by='Symbol').reset_index(drop=True)

    perc_df = df1.merge(df2, how='inner', left_on='Symbol', right_on='Symbol')
    perc_df['per_change'] = round((perc_df.Price - perc_df.Yesterday)/perc_df.Yesterday *100,2)
    perc_df.drop(columns='Price', inplace=True)
    
    df_int = df_all.merge(perc_df, how='outer', left_on='Symbol', right_on='Symbol')
    
    return df_int, perc_df
```

```python
df_int, perc_df = concat_df(df_all, new)
```

```python
def calculations(df_final):
    
    
    put_call_df = pd.DataFrame(columns=['Symbol', 'Calls', 'Puts', 'Put/Call', 'TotalVolume', 'Orient'])

    for symbol in df_final.Symbol.unique():

        df = df_final[df_final.Symbol == symbol]
        calls = df[df.Type == 'Call'].Volume.sum()
        puts = df[df.Type == 'Put'].Volume.sum()
        TotalVol = df.Volume.sum()
        

        put_call = puts/calls

        inner_df = pd.DataFrame(columns=['Symbol', 'Calls', 'Puts', 'Put/Call', 'TotalVolume', 'Orient'])
        inner_df['Symbol'], inner_df['Calls'], inner_df['Puts'], inner_df['Put/Call'], inner_df['TotalVolume'] = [symbol], [calls], [puts], [put_call], [TotalVol]
        
        if inner_df.Calls.values > inner_df.Puts.values:
            inner_df['Orient'] = 'Calls'
        elif inner_df.Calls.values < inner_df.Puts.values:
            inner_df['Orient'] = 'Puts'
        elif inner_df.Calls.values == inner_df.Puts.values:
            inner_df['Orient'] = 'Tie'

        put_call_df = pd.concat([put_call_df, inner_df])

    put_call_df = put_call_df.reset_index(drop='True')
    put_call_df = put_call_df.sort_values(by='Symbol')

    change_cal_df = pd.merge(perc_df, put_call_df, how='inner', left_on='Symbol', right_on='Symbol')
    change_cal_df = change_cal_df[(change_cal_df['Put/Call'] <= 0.50) & (change_cal_df['Put/Call'] <= 1.4)| (change_cal_df['Put/Call'] >= 1.5) | (change_cal_df['Put/Call'] == np.inf)]
    change_cal_df.sort_values(by='Put/Call', ascending=False)
    change_cal_df['Time'] = df_all.Time.max().strftime('%Y-%m-%d') # Fix this 

    final_cal_df = change_cal_df[((change_cal_df.Calls >= 10000) | (change_cal_df.Puts >= 10000)) & (abs(change_cal_df.per_change) <= 5 )]
    final_cal_df['Put/Call'] = final_cal_df['Put/Call'].replace(np.inf, 99.99)
    final_cal_df['record_insert_time'] = pd.datetime.now().strftime('%m-%d %H:%M')

    return final_cal_df, change_cal_df, put_call_df
```

```python
final_cal_df, change_cal_df, put_call_df = calculations(df_int)
```
