```python
def active_options_data(df):
    
    df = df[['Symbol', '%Chg', 'IV_Rank', 'Options_Vol', '%_Put', '%_Call',  'Put/Call_Vol', 'Time']]

    app_df = pd.DataFrame(columns=['Symbol', '%Chg', 'IV_Rank', 'Options_Vol', '%_Put', '%_Call',  'Put/Call_Vol', 'Repeat'])
    for sym in df.Symbol.unique():

        inn_df = pd.DataFrame(columns=['Symbol', '%Chg', 'IV_Rank', 'Options_Vol', '%_Put', '%_Call',  'Put/Call_Vol', 'Repeat'])

        df_ind = df[df.Symbol == sym]
        inn_df.Symbol, inn_df['%Chg'], inn_df['IV_Rank'] = [sym], [df_ind['%Chg'].mean()], [df_ind['IV_Rank'].mean()]
        inn_df['Options_Vol'], inn_df['%_Put'], inn_df['%_Call'], inn_df['Put/Call_Vol'] = [df_ind['Options_Vol'].mean()], [df_ind['%_Put'].mean()], [df_ind['%_Call'].mean()], [df_ind['Put/Call_Vol'].mean()]
        inn_df['Repeat'] = len(df_ind)
        app_df = pd.concat([inn_df, app_df])

    app_df.reset_index(drop=True, inplace=True)

    app_df = app_df[app_df.Symbol != 'BRK.B']
    
    return app_df
```

```python
#Trying without manipulating bid and ask price but try later with bid and ask above
uao = pd.read_sql_query('SELECT * FROM options.unusual_stock_options_activity WHERE Time >= "2022-05-31" AND DTE <=90;',conn)
```

```python
def uoa_data_prep(uao):

    cols=['Symbol', 'put_dte', 'call_dte', 'put_uao_repeat', 'call_uao_repeat',
                                    'call_bid', 'call_mid', 'call_ask', 'call_last',
                                    'put_bid', 'put_bid', 'put_bid', 'put_bid',
                                    'call_price', 'put_price', 'call_vol', 
                                    'put_vol', 'call_iv', 'put_iv', 'call_delta', 'put_delta']

    outer_df = pd.DataFrame(columns=cols)

    for sym in uao.Symbol.unique():

    #     inner_df = pd.DataFrame(columns=['Symbol', 'put_dte', 'call_dte', 'put_uao_repeat', 'call_uao_repeat',
    #                                     'call_bid', 'call_mid', 'call_ask', 'call_last',
    #                                     'put_bid', 'put_bid', 'put_bid', 'put_bid'])



        uao_ind = uao[uao.Symbol == sym]

        put_df = uao_ind[uao_ind.Type == 'Put']
        call_df = uao_ind[uao_ind.Type == 'Call']

        put_dte = put_df.DTE.mean()
        call_dte = call_df.DTE.mean()

        put_uao_repeat = len(put_df)
        call_uao_repeat = len(call_df)

        put_strike = put_df.Strike.mean()
        call_strike = call_df.Strike.mean()

        call_bid, call_mid, call_ask, call_last = call_df.Bid.mean(), call_df.Midpoint.mean(), call_df.Ask.mean(), call_df.Last.mean()
        put_bid, put_bid, put_bid, put_bid = put_df.Bid.mean(), put_df.Midpoint.mean(), put_df.Ask.mean(), put_df.Last.mean()

        call_price = call_df.Price.max() - call_df.Price.min()
        put_price = put_df.Price.max() - put_df.Price.min()

        call_vol = call_df.Volume.sum()
        put_vol = put_df.Volume.sum()

        call_oi = call_df.Open_Int.sum()
        put_oi = put_df.Open_Int.sum()

        call_iv = call_df.IV.mean()
        put_iv = put_df.IV.mean()

        call_delta = call_df.Delta.mean()
        put_delta = put_df.Delta.mean()



        inner_df = pd.DataFrame([[sym, put_dte, call_dte, put_uao_repeat, call_uao_repeat,
                                 call_bid, call_mid, call_ask, call_last,
                                 put_bid, put_bid, put_bid, put_bid, 
                                 call_price, put_price, call_vol, put_vol, call_iv, put_iv, call_delta, put_delta]], columns = cols)

        outer_df = pd.concat([inner_df, outer_df])

    outer_df = round((outer_df.fillna(0)), 2)
    outer_df.reset_index(drop=True, inplace=True)

    outer_df = outer_df[outer_df.Symbol != 'BRK.B']
    
    return outer_df
```
